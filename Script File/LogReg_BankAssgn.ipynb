```python
#one hot encoding
```

```python
data=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
pd.set_option('display.max.columns', None) #display all columns
data
```

```python
data.info()
```

```python
data.default.replace(('yes', 'no'), (1, 0), inplace=True)
data.loan.replace(('yes', 'no'), (1, 0), inplace=True)
data.y.replace(('yes', 'no'), (1, 0), inplace=True)
data.housing.replace(('yes', 'no'), (1, 0), inplace=True)
```

```python
data
```

```python
data['month'].value_counts()
```

```python
mon_order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
data_n=data.replace(mon_order)
```

```python
data_n
```

```python
data_n.info()
```

```python
# Dividing our data into input and output variables 
X = pd.concat([data_n.iloc[:,0:11],data_n.iloc[:,12:]],axis=1)
Y = data_n.iloc[:,11]
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression(solver='liblinear')
classifier.fit(X,Y)
```

```python
#Predict for X dataset
y_pred = classifier.predict(X)
```

```python
y_pred_df= pd.DataFrame({'actual': Y,
                         'predicted_prob': y_pred})
```

```python
y_pred_df
```

```python
# Confusion Matrix for the model accuracy
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y,y_pred)
print (confusion_matrix)
```

```python
#Classification report
from sklearn.metrics import classification_report
print(classification_report(Y,y_pred))
```

```python
# ROC Curve
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score

fpr, tpr, thresholds = roc_curve(Y, classifier.predict_proba (X)[:,1])

auc = roc_auc_score(Y, y_pred)

import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red', label='logit model ( area  = %0.2f)'%auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
